# Imports

```python
import numpy as np
from keras.models import Sequential
from keras.layers.core import Dense
from keras.wrappers.scikit_learn import KerasClassifier
import matplotlib.pyplot as plt
from random import shuffle
from sklearn.model_selection import StratifiedKFold, GridSearchCV, ParameterGrid

# fix random seed for reproducibility
seed = 7
```

# Tutorial

**1) Data:**

```python
# the four different states of the XOR gate
training_data = np.array([[0,0],[0,1],[1,0],[1,1]], "float32")

# the four expected results in the same order
target_data = np.array([[0],[1],[1],[0]], "float32")
```

**2) Create model:**

```python
model = Sequential()
model.add(Dense(16, input_dim=2, activation='relu'))
model.add(Dense(1, activation='sigmoid'))

model.compile(loss='mean_squared_error',
              optimizer='adam',
              metrics=['binary_accuracy'])
```

**3) Run model:**

```python
model.fit(training_data, target_data, nb_epoch=500, verbose=2)
```

```python
print(model.predict(training_data).round())
```

# XOR - (b)

### Create Data:

```python
# Generates group 1
g1 = np.column_stack((np.random.normal(0, 0.1, 200), np.random.normal(0, 0.1, 200)))

# Generates group 2
g2 = np.column_stack((np.random.normal(0, 0.1, 200), np.random.normal(1, 0.1, 200)))

# Generates group 3
g3 = np.column_stack((np.random.normal(1, 0.1, 200), np.random.normal(0, 0.1, 200)))

# Generates group 4
g4 = np.column_stack((np.random.normal(1, 0.1, 200), np.random.normal(1, 0.1, 200)))

# Features
X = np.concatenate((g1,g2, g3, g4), axis = 0)

# Target
y = np.concatenate((np.zeros(200), np.ones(400), np.zeros(200)), axis = 0)
```

```python
fig, ax = plt.subplots()
ax.plot(X[:, 0], X[:, 1], '.')
fig.show()
```

## Split data

```python
# Total Data in Tuple Format
data = list(zip(X, y))
```

```python
size= len(data)

# shuffle data vector
shuffle(data)

# select 80% for training
train = data[0:round(0.8*size)]

#select 20% for test
test = data[round(0.8*size):size]
```

## Create Model

```python
def model_creator(activation_per_layer, input_dim, loss_fct, metrics, neurons_per_layer, optimizer):
    '''
    :param: input_dim - input dimension
    :param neurons_per_layer - list of neurons in each layer
    :param activation_per_layer - list of activation functions in each layer
    :param loss_fct - loss function to use
    :param optimizer - type of optimizer
    :param metrics - measuring metrics
    '''
    model = Sequential()
    for i in range(len(neurons_per_layer)):
        if i == 0:
            model.add(Dense(neurons_per_layer[i], input_dim = input_dim, activation=activation_per_layer[i]))
        else:
            model.add(Dense(neurons_per_layer[i], activation = activation_per_layer[i]))

    model.compile(loss= loss_fct, optimizer=optimizer, metrics=metrics)
    
    return model
```

## Fit model

```python
train_X = np.asarray([i[0] for i in train])
train_target = np.asarray([i[1] for i in train])
```

### Performing Cross Validation on grid of parameters...

**Grid Parameters:**

```python
neurons = [[16, 1], [32, 1], [64, 1], [128,1]] # number of nodes in the hidden layer
activations = [['relu', 'sigmoid']]
epochs = 500

param_grid = dict(input_dim = [2], neurons_per_layer=neurons, activation_per_layer = activations,
                  loss_fct = ['mean_squared_error'], optimizer = ['sgd'], metrics = [['accuracy']])
```

```python
model = KerasClassifier(build_fn=model_creator, epochs=epochs, verbose=0)

grid = GridSearchCV(estimator=model, param_grid=param_grid, cv=3,
                    n_jobs=1, refit=True, verbose=2)
grid_result = grid.fit(train_X, train_target)
```

# Performance

```python
def performance_evaluation(model, test):
    test_X = np.asarray([i[0] for i in test])
    test_target = np.asarray([i[1] for i in test])

    # evaluation of model on test data
    score = model.evaluate(test_X, test_target, verbose=1)
    print('Final Score: ', score)
```

```python
performance_evaluation(grid_result, test)
```

```python
test_X = np.asarray([i[0] for i in test])
test_target = np.asarray([i[1] for i in test])

grid_result.error_score(test_X, test_target)
```
